```python
####  educacion
import pandas as pd
from pandasql import sqldf
mysql = lambda q: sqldf(q, globals())
df=pd.read_csv('noticias_derechos10.csv')
df = mysql("""SELECT * FROM df WHERE ((date) BETWEEN '2021-07-04' AND '2022-07-27') AND (LOWER(body) LIKE LOWER ('%ConvenciÃ³n%') AND LOWER(body) LIKE LOWER ('%Constitu%') AND body LIKE '%derecho%' AND body LIKE '%educaciÃ³n%'
OR LOWER(body) LIKE LOWER ('%Nueva%') AND LOWER(body) LIKE LOWER ('%ConstituciÃ³n%') AND body LIKE '%derecho%'  AND body LIKE '%educaciÃ³n%')
ORDER BY date""")
df
```

```python
###### Numero de medios 
medios = df['title_1'].unique()
medios = pd.DataFrame(medios)
medios
```

```python
#### conteo de medios 
freq2 = df['title_1'].value_counts() 
conteo_medios = pd.DataFrame(freq2) 
conteo_medios
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df_xsemana = df.assign(N_noticias=1) ## crear columna con 1
df_xsemana = df_xsemana.loc[:,['date','N_noticias']] ##### seleccion de columnas para df
df_xsemana = df_xsemana.groupby(pd.Grouper(key='date', axis=0, 
                      freq='7D', sort=True)).sum()          ##### agrupa x semana y suma
df_xsemana
```

```python
df_xsemana.to_csv('educacion_noticias_xsemanas.csv')
files.download('educacion_noticias_xsemanas.csv')
```
